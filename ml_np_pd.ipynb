##structured data and unstructured data



### Labeled and Unlabeled Data

**Labeled Data:**
- **Definition:** Data that includes labels or tags identifying characteristics, properties, or classifications.
- **Example:** A dataset with images of animals where each image is tagged with the animal type (e.g., cat, dog, bird).

**Unlabeled Data:**
- **Definition:** Data that does not have labels identifying characteristics, properties, or classifications.
- **Example:** A dataset with images of animals where no information about the type of animal in each image is provided.

### Types of Data: Static and Streaming

**Static Data:**
- **Definition:** Data that does not change once it is recorded.
- **Example:** Historical sales data that remains constant over time.

**Streaming Data:**
- **Definition:** Data that is continuously generated and changes over time.
- **Example:** Real-time sensor data from a weather station.

### Types of Variables

**Numerical Data (Numerical):**
- **Discrete Data:**
  - **Definition:** Data that can only take specific values (usually whole numbers).
  - **Example:** Number of students in a class.
- **Continuous Data:**
  - **Definition:** Data that can take any value within a given range.
  - **Example:** Height of students.

**Categorical Data (String Data):**
- **Nominal Data:**
  - **Definition:** Categories without a particular order.
  - **Example:** Types of animals (e.g., cat, dog, bird).
- **Ordinal Data:**
  - **Definition:** Categories with a natural order or ranking.
  - **Example:** Educational levels (e.g., high school, bachelor's, master's).

**Datetime Variable:**
- **Definition:** Variables that represent dates and times.
- **Example:** A timestamp indicating when a transaction occurred.

### Categorical Data

**Nominal Data:**
- Think of it like different labels with no particular order.
- Example: Colors (red, blue, green) or types of animals (cat, dog, bird).

**Ordinal Data:**
- Categories with a natural order or ranking, but the differences between them might not be equal.
- Example: Educational levels (high school, bachelor's, master's) or Likert scale responses (strongly disagree, disagree, neutral, agree, strongly agree).

In this version, we use __init__ to initialize the name attribute during object creation

```python
class PersonWithInit:
    def __init__(self, name):
        self.name = name

# Creating instances with names
person1 = PersonWithInit("Alice")
person2 = PersonWithInit("Bob")

# Accidentally overwriting name attribute of person1
person1.name = "Charlie"

print("Name of person1:", person1.name)  # Output: Charlie
print("Name of person2:", person2.name)  # Output: Bob
```

no __init__ method, so we manually set the name attribute using the set_name method

```python
class PersonWithoutInit:
    def set_name(self, name):
        self.name = name

# Creating instances and setting names
person1 = PersonWithoutInit()
person2 = PersonWithoutInit()

person1.set_name("Alice")
person2.set_name("Bob")

# Accidentally overwriting name attribute of person1
person1.name = "Charlie"

print("Name of person1:", person1.name)  # Output: Charlie
print("Name of person2:", person2.name)  # Output: Bob
```

In computer science and linear algebra, a vector can be represented as either a row vector or a column vector, depending on the context.

1. **Row Vector**:
   - A row vector is a one-dimensional array of numbers arranged horizontally.
   - It is often denoted as a matrix with a single row.
   - Example: The vector $ \mathbf{v} = [3, -2, 5] $ is a row vector.


2. **Column Vector**:
   - A column vector is a one-dimensional array of numbers arranged vertically.
   - It is often denoted as a matrix with a single column.
   - Example: The vector $ \mathbf{v} = \begin{bmatrix} 3 \\ -2 \\ 5 \end{bmatrix} $ is a column vector.


Both representations are valid and equivalent. The choice between row and column vectors depends on the application and convenience. In many contexts, column vectors are more commonly used, especially in linear algebra and machine learning, but row vectors are also used, particularly in contexts like transformations and geometric operations.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
smark=[80,90,70,65,85,90,95]
```

```python

import numpy as np
sroll=[13,15,21,67,46,62,38]
sroll=np.sort(sroll)
```

```python
plt.plot(sroll,smark)#x,y
plt.xlabel("rollnumber")
plt.ylabel("marks")
plt.title("marks of students")
```

```python
##ploting using seaborn
```

```python
sns.lineplot(x=sroll,y=smark)
```

```python
plt.bar(sroll,smark)
plt.show()#ploting the bar
```

```python
#if we want to add colors to bars
c=['r','y','b','g','c','b']
plt.bar(sroll,smark,color=c)
plt.xlabel("roll number")
plt.ylabel("marks")
```

```python

Capacity=[300,700,178,600,250,240,180]
plt.scatter(sroll,smark,marker="3",color='y',s=Capacity,alpha=1)#scatter points on graphs color = color,marker =point shape s=marker capacity
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming smark and Capacity are defined

# Define the bin size
bin_size = 0.1  # Specify the size of each bin

# Calculate the range of values
min_val = np.min(smark)
max_val = np.max(smark)
bins = np.arange(min_val, max_val + bin_size, bin_size)  # Create bins using specified bin size

# Calculate histogram bins using np.histogram
hist, _ = np.histogram(smark, bins=bins)

# Plot the histogram
plt.hist(smark, bins=bins)

# Show plot
plt.show()
```

```python
plt.pie(smark,labels=sroll,shadow=True,explode=[1,0,1,0,0,0,0])#values,labels,shadow,
plt.legend(title="total marks by students")#piechart
```

```python
import matplotlib.pyplot as plt

# Format labels to include marks
labels_with_marks = [f"{label} ({mark})" for label, mark in zip(sroll, smark)]

plt.pie(smark, labels=labels_with_marks, shadow=True, explode=[0.1, 0, 0, 0, 0, 0, 0])  # Values, labels with marks, shadow, explode
plt.legend(title="Total marks by students")  # Pie chart with values
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the labels
labels = sroll

# Plot the pie chart with autopct to display the numeric values
plt.pie(smark, labels=labels, shadow=True, explode=[0.1, 0, 0, 0, 0, 0, 0], autopct='%1.1f%%')
plt.legend(title="Total marks by students")
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Create a DataFrame with smark as the data and sroll as the index
data = pd.DataFrame({'Marks': smark}, index=sroll)

# Create the heatmap
sns.heatmap(data, annot=True, fmt=".1f", cmap="YlGnBu")

# Set labels and title
plt.xlabel('Marks')
plt.ylabel('Student Roll Numbers')
plt.title('Heatmap of Marks by Student')

# Show the plot
plt.show()
```

```python
##multiplication vector by scaler
h=np.asarray([0,0,2,3])
v=2
mul=h*v
print(mul)


plt.quiver(0,0,4,6,scale_units="xy",angles="xy",scale=1,color="b")#base point,head point#scale unit =1
plt.quiver(0,0,2,3,scale_units="xy",angles="xy",scale=1,color="y")#orginal vector
plt.xlim(-8,8)#xlimit,ylimit
plt.ylim(-8,8)
plt.show()
```
